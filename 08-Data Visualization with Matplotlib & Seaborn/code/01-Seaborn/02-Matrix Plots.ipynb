# Matrix Plots

* Heatmap

Matrix plots allow you to plot data as color-encoded matrices and can also be used to indicate clusters within the data (later in the machine learning section we will learn how to formally cluster data).

Let's begin by exploring seaborn's heatmap and clutermap:

```python
import seaborn as sns
%matplotlib inline
sns.set(rc={'figure.figsize': [15, 15]}, font_scale=1.3)
```

```python
flights = sns.load_dataset('flights')
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
flights
```

## Heatmap

In order for a heatmap to work properly, your data should already be in a matrix form, the sns.heatmap function basically just colors it in for you. For example:

```python
# Matrix form for correlation data
tips.corr()
```

```python
sns.heatmap(tips.corr())
```

```python
sns.heatmap(tips.corr(), annot=True)
```

Or for the flights data:

```python
flights
```

```python
flights = flights.pivot_table(values='passengers', index='month', columns='year')
flights
```

```python
sns.heatmap(flights, annot=True)
```

**we can customize the seperating line's width and color**

```python
sns.heatmap(flights, linecolor='white', linewidths=2)
```

or you can use this awesome **cubehelix_palette**

```python
cmap = sns.cubehelix_palette(as_cmap=True, reverse=True, dark=0, light=1)
sns.heatmap(flights, cmap=cmap)
```

```python
cmap = sns.cubehelix_palette(as_cmap=True, reverse=True, dark=0, light=1)
sns.heatmap(flights, linecolor='k', linewidths=2, cmap=cmap)
```
